```python
import pandas as pd

file_path = '../data/benin-malanville.csv'
df = pd.read_csv(file_path)

print(df.head())
```

# Summary Statistics


```python
# Summary statistics for numeric columns
summary_stats = df.describe()
print(summary_stats)

# Additional statistics
median_values = df.median(numeric_only=True)
print("\nMedian Values:\n", median_values)
```

# Data Quality Check


```python
# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values)

# Check for negative values in GHI, DNI, DHI
negative_values = df[(df['GHI'] < 0) | (df['DNI'] < 0) | (df['DHI'] < 0)]
print("\nRows with Negative Values:\n", negative_values)

# Detect outliers using the Z-score method
from scipy.stats import zscore

outliers = df[(zscore(df[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']]) > 3).any(axis=1)]
print("\nPotential Outliers:\n", outliers)
```

# Time Series Analysis

```python
import matplotlib.pyplot as plt

# Convert Timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Plot time series
plt.figure(figsize=(12, 6))
for col in ['GHI', 'DNI', 'DHI', 'Tamb']:
    plt.plot(df['Timestamp'], df[col], label=col)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series of GHI, DNI, DHI, and Tamb')
plt.legend()
plt.show()
```

# Impact of Cleaning

```python
# Compare ModA and ModB values when Cleaning = 1 vs. 0
cleaned = df[df['Cleaning'] == 1]
uncleaned = df[df['Cleaning'] == 0]

print("\nImpact of Cleaning on ModA:")
print("Mean (Cleaned):", cleaned['ModA'].mean())
print("Mean (Uncleaned):", uncleaned['ModA'].mean())
```

# Correlation Analysis

```python
import seaborn as sns

# Correlation matrix
correlation_matrix = df[['GHI', 'DNI', 'DHI', 'TModA', 'TModB', 'WS', 'WSgust']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot: RH vs Tamb
plt.scatter(df['RH'], df['Tamb'])
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Temperature (C)')
plt.title('Relationship Between RH and Temperature')
plt.show()
```

# Histograms


```python
# Histograms for selected variables
df[['GHI', 'DNI', 'DHI', 'WS', 'Tamb']].hist(bins=20, figsize=(12, 8))
plt.suptitle('Histograms of GHI, DNI, DHI, WS, and Tamb')
plt.show()
```

# Z-Score Analysis

```python
# Z-scores for solar radiation components
z_scores = zscore(df[['GHI', 'DNI', 'DHI']])
df['Z-score Flag'] = (abs(z_scores) > 3).any(axis=1)
print("\nRows Flagged with Z-scores > 3:\n", df[df['Z-score Flag']])
```

# Bubble Charts

```python
# Bubble chart: GHI vs Tamb vs WS, bubble size = RH
plt.scatter(df['GHI'], df['Tamb'], s=df['RH']*10, alpha=0.5, c=df['WS'], cmap='viridis')
plt.colorbar(label='Wind Speed (m/s)')
plt.xlabel('GHI')
plt.ylabel('Tamb')
plt.title('GHI vs Tamb vs WS (Bubble size = RH)')
plt.show()
```

# Data Cleaning

```python
# Drop rows where Comments are entirely NaN
df_cleaned = df.dropna(subset=['Comments'])

# Replace negative values with NaN in GHI, DNI, DHI
df_cleaned[['GHI', 'DNI', 'DHI']] = df_cleaned[['GHI', 'DNI', 'DHI']].applymap(lambda x: x if x >= 0 else None)

# Display the cleaned dataset
print(df_cleaned.head())
```
