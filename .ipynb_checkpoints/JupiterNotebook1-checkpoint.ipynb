#  Z-Score Analysis


```python
from scipy.stats import zscore

numeric_cols = data.select_dtypes(include='number').columns
z_scores = data[numeric_cols].apply(zscore)
outliers = (z_scores.abs() > 3).sum()
print(outliers)
```

#  Time Series Analysis

```python
import matplotlib.pyplot as plt

data['Timestamp'] = pd.to_datetime(data['Timestamp'])
plt.figure(figsize=(10, 6))
plt.plot(data['Timestamp'], data['GHI'], label='GHI')
plt.plot(data['Timestamp'], data['DNI'], label='DNI')
plt.plot(data['Timestamp'], data['DHI'], label='DHI')
plt.legend()
plt.show()
```

# Bubble charts


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm


fig, ax = plt.subplots(figsize=(10, 6))


sc = ax.scatter(
    data['GHI'],    
    data['Tamb'],   
    s=data['RH'],   
    c=data['BP'],  
    cmap='viridis', 
    alpha=0.7      
)


cbar = plt.colorbar(sc)
cbar.set_label('BP (Barometric Pressure)')


ax.set_xlabel('GHI (Global Horizontal Irradiance)')
ax.set_ylabel('Tamb (Ambient Temperature)')
ax.set_title('Bubble Chart of GHI vs. Tamb vs. RH/BP')

plt.show()
```
